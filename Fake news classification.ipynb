```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
data_title['title_cleaned'] = data_title['title_cleaned'].apply(
    lambda x : " ".join(
        x for x in x.split() if x not in stop
    )
)

data_title['title_cleaned']
```

```python
tokenized = []
for i in range(len(data_title['title_cleaned'])):
    tokenized.append(nltk.word_tokenize(data_title['title_cleaned'][i]))
    
tokenized
```

```python
data_title['title_cleaned'] =  data_title['title_cleaned'].apply(lambda x : ' '.join(i for i in x))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_title['title_cleaned'],
                                                    data_title['label'], 
                                                    test_size = 0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
V = CountVectorizer()
cv = V.fit_transform(X_train)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(cv, y_train)
```

```python
X_test_cv = V.transform(X_test)
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_cv)
print(classification_report(y_test, y_pred))
```

RNN

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import SimpleRNN, LSTM, GRU
from tensorflow.keras.layers import Dense
```

```python
voc_size = 5000
```

```python
messages = data_title['title_cleaned'].copy()
messages = messages.reset_index()
```

```python
y = data_title['label']
```

```python
corpus = []
for i in range(messages.shape[0]):
    corpus.append(messages['title_cleaned'][i])
```

```python
def word_embeddings(corpus):
    onehot_repr = [one_hot(words, voc_size) for words in corpus]
    sent_length = 50
    embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen = sent_length)
    return embedded_docs
```

```python
embedded_docs = word_embeddings(corpus)
```

```python
embedding_vector_features = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_vector_features, input_length = sent_length))
model.add(LSTM(150))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.summary()
```

```python
len(embedded_docs), y.shape
```

```python
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
X_final.shape, y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.2, 
                                                    random_state = 42)
```

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 10, batch_size = 64)
```

```python
y_pred = model.predict(X_test)
y_pred = np.where(y_pred > 0.5, 1, 0)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
